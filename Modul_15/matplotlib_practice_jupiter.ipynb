# Ćwiczenie

Stwórz wykres z płacami zorientowany poziomo. Niech kolory poziomych kolumn będą zielone. Wynik ma być podobny do wykresu poniżej.

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
```

```python
plt.title("Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary in zł")
plt.bar(names, salary_values)
```

```python
import matplotlib.pyplot as plt

y_axis = names
x_axis = salary_values

plt.barh(y_axis,x_axis)
plt.title('Salaries')
plt.ylabel('Employee')
plt.xlabel('Salary in zł')
plt.show()
```
